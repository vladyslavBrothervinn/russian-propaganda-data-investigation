### 5. What is the distribution of posts across months? Are there any seasonal trends?

```python
df_q1_v2.groupby("month").count()
df_temp = pd.crosstab(index=df_q1_v2["month"], columns='post_quantity')
df_temp['month_name'] = ['February', 'March', 'April', 'May', 'June', 'July', 'August', 'September',
                         'October', 'November', 'December']
plt.style.use('seaborn')

# Let's find the mean posts quantity per month and mark it on the graph
mean_val = df_temp['post_quantity'].mean()

plt.figure(figsize = (13, 10))
df_temp.plot(kind = 'bar', x = 'month_name', y = 'post_quantity', legend = False)

plt.title('Posts distribution per month')
plt.xlabel('Month', color = 'black')
plt.ylabel('Number of Posts', color = 'black')
plt.xticks(rotation = 45, ha = 'right', color = 'black')

plt.axhline(mean_val, color='red', linestyle='--', linewidth=3)
plt.text(5, mean_val, f'Mean Value', color='black', va='bottom', ha='center')

plt.show()
```

As observed, there are 6 months that surpass the mean threshold value for post activity, with the highest post amounts recorded in March, April, and October, respectively. Notably, these peaks coincide with significant events that likely influenced the surge in post activity.

In March, there was heightened activity due to the commencement of warfare, particularly in the Kyiv region, with notable fights in Bucha, Irpin, and Hostomel, which may explain the increased posts during this period.

Continuing into April and May, conflicts persisted in the East and South of Ukraine, culminating in the extended battle in Mariupol, which garnered substantial attention and continued through May. The conclusion of this conflict, particularly impactful for Russia, likely contributed to a surge in posts during this timeframe.

Moreover, at the end of September, Russia announced mobilization, a significant event that could have provoked the high post flow observed during that month.

### 10. What are the most frequently used words in the posts throughout the warfare.

#### Let's take the most active channels, swodki, karaulny and glavmedia, and based on their posts investigate the question.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# nltk.download('stopwords')
# nltk.download('punkt')

def extract_keywords_with_tf_idf_vectors(text, stop_words):
    if isinstance(text, str):  # Check if the entry is a string
        tokens = word_tokenize(text, language='russian', preserve_line=True)
        filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
        return ' '.join(filtered_tokens)  # Modified to return a string of tokens
    else:
        return ''

stop_words = set(stopwords.words('russian'))

df_q1_v2['keywords_tf-idf'] = df_q1_v2['message'].apply(lambda x: extract_keywords_with_tf_idf_vectors(x, stop_words))

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df_q1_v2['keywords_tf-idf'])

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

print(tfidf_df)
```

```python
from collections import Counter

def most_frequent_words(df):
    word_count = Counter()
    for row in df:
        words = row.split()
        for word in words:
            for suffix_length in range(1, 5):
                suffix = word[:-suffix_length]
                if suffix in word_count:
                    word_count[word] += 1
                    break
                else:
                    word_count[word] += 1
    
    return word_count
```

```python
def dict_to_df(d):
    df = pd.DataFrame(list(d.items()), columns=['word', 'count'])
    df['len'] = df['word'].apply(lambda x: len(x))  
    mask = df['len'] > 2 
    df = df[mask]
    df = df.sort_values(by='count',ascending=False)
    df = df.set_index(pd.Index([i for i in range(1, df.shape[0]+1)]))
    return df
```

```python
import circlify
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

def to_graph(df, title, s):

    circles = circlify.circlify(
        df['count'].tolist(), 
        show_enclosure=False, 
        target_enclosure=circlify.Circle(x=0, y=0, r=1)
    )
    circles.reverse()

    fig, ax = plt.subplots(figsize=(10, 10))

    ax.set_title(title)
    ax.axis('off')

    lim = max(
        max(
            abs(circle.x) + circle.r,
            abs(circle.y) + circle.r,
        )
        for circle in circles
    )
    plt.xlim(-lim, lim)
    plt.ylim(-lim, lim)

    words = [f"{df.iloc[i, 0]}\n" for i in range(df.shape[0])]
    counts = [df.iloc[i, 1] for i in range(df.shape[0])]  # Extracting values from the set
    
    color_list = list(sns.color_palette(s, n_colors=5).as_hex())
    
    counter=0
    i=-1
    
    plt.style.use('seaborn')
    
    for circle, word, count in zip(circles, words, counts):
        if counter%15==0:
            i+=1
        x, y, r = circle
        # Adjust the radius based on the count
        #scaled_r = r * (count / max(counts))*2
        ax.add_patch(plt.Circle((x, y), r, color=color_list[i], alpha=1))
        plt.annotate(
            word + str(count), 
            (x, y),
            va='center',
            ha='center',
            fontsize=11
         )
        counter+=1
    plt.show()
```
